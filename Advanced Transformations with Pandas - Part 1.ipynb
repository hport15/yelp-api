# Separate a string column into multiple columns

## Using .str.split with expand=True to create multiple columns

```python
# explore existing format
df['display_phone'].head(2)
```

```python
# add expand=True automatically splits col into multiple cols
df['display_phone'].str.split(' ', expand=True)
```

```python
# save the 2 new cols in df
df[['area_code', 'phone_number']] = df['display_phone'].str.split(' ',
                                                                  expand=True)
df.head(2)
```

```python
# drop original column
df.drop(['display_phone','phone'], axis=1, inplace=True)
df.head(2)
```

# Replacing multiple characters at once within a string column

## Using df[col].str.replace in a loop

```python
# list of characters to "replace" (really remove)
to_replace = ['(', ')']
# loop to remove the characters
for char in to_replace:
    df['area_code'] = df['area_code'].str.replace(char, "", regex=False)
df['area_code'].head(2)
```

# Converting a string column of dictionaries into actual dictionaries.

## Testing Our Approach with a Single Value

```python
# examine single value from coordinates col
coord = df.loc[0,"coordinates"]
print(type(coord))
coord
```

"json.loads" (the s is for "strings"),loads a JSON-compatible STRING into a dictionary/list of dictionaries.

```python
import json
json.loads(coord)
```

```python
# must use "" not '' for json. replace '' with ""
coord = coord.replace("'",'"')
coord
```

```python
# run json.loads again
fixed_coord = json.loads(coord)
print(type(fixed_coord))
fixed_coord
```

```python
# apply to the entire column
# use .str.replace to replace all '
df['coordinates'] = df['coordinates'].str.replace("'",'"')
# apply json.loads to entire col
df['coordinates'] = df['coordinates'].apply(json.loads)
df['coordinates'].head(2)
```

# Unpack a column of dictionaries into separate columns

```python
# separate coordinates column into 2 using pd.Series
lat_long = df['coordinates'].apply(pd.Series)
lat_long
```

```python
# add the 2 new cols to the df and drop the original
df = pd.concat((df, lat_long), axis=1)
df.head(2)
```

```python
# drop coordinates col
df.drop('coordinates', axis=1, inplace=True)
df.head(2)
```

# Convert a column that contains lists (that were converted to strings), back to lists.

## Getting All Unique Options (Columns to Make)

```python
# display the "lists" that are actually str
df.loc[2, 'transactions']
```

```python
# create new col where the single quotes in the list are replaced with "
df['transactions_split'] = df['transactions'].str.replace("'", '"')
```

```python
# apply json.loads to convert to list
df['transactions_split'] = df['transactions_split'].apply(json.loads)
df['transactions_split'].head()
```

# Convert a column of lists into one-hot-encoded columns (a column for each item in the list)

```python
df['transactions_split'].value_counts()
```

## Get a list of all options using df.explode

```python
# use df.explode to create a new row for each value in the list
exploded = df.explode('transactions_split')
exploded[['name', 'transactions', 'transactions_split']].head(5)
```

```python
# save unique values from exploded col
cols_to_make = exploded['transactions_split'].dropna().unique()
cols_to_make
```

## Using a For Loop and .str.contains to create the new columns

```python
for col in cols_to_make:
    df[col] = df['transactions'].str.contains(col)
df.head()
```
